```python
#Number of male and female participants per group
count_c_m = grouped_df['sex'].value_counts()[0][0]
count_c_f = grouped_df['sex'].value_counts()[0][1]

counts_sex_c = "%d/%d" % (count_c_m, count_c_f)

count_p_m = grouped_df['sex'].value_counts()[2][0]
count_p_f = grouped_df['sex'].value_counts()[2][1]

counts_sex_p = "%d/%d" % (count_p_m, count_p_f)
```

```python
#Subject age
avg_c_age = grouped_df['age_Y'].mean()[0]
std_c_age = grouped_df['age_Y'].std()[0]

avg_p_age = grouped_df['age_Y'].mean()[2]
std_p_age = grouped_df['age_Y'].std()[2]
```

```python
control_age_string = "%2.3f %s %2.3f" % (round(avg_c_age, 3), u"\u00B1", 2*round(std_c_age, 3))
patient_age_string = "%2.3f %s %2.3f" % (round(avg_p_age, 3), u"\u00B1", 2*round(std_p_age, 3))
```

```python
#Mother's education scale
mot_c_avg = controls['mot_edscale'].dropna().mean()
mot_c_std = controls['mot_edscale'].dropna().std()

mot_p_avg = n_affective['mot_edscale'].dropna().mean()
mot_p_std = n_affective['mot_edscale'].dropna().std()

mot_c_str = "%3.2f %s %1.2f" % (controls['mot_edscale'].mean(), u"\u00B1", 2*controls['mot_edscale'].std())
mot_p_str = "%3.2f %s %1.2f" % (mot_p_avg, u"\u00B1", 2*mot_p_std)
```

```python
#AUDITORY RESPONSE TIMES and HIT RATES (Averages and standard deviations stored in strings for demographics table)
avg_c_comp_aud_rt = controls['comp_aud_RT'].mean()
std_c_comp_aud_rt = controls['comp_aud_RT'].std()

c_comp_aud_rt_str = "%3.2f %s %3.2f (ms)" % (avg_c_comp_aud_rt, u"\u00B1", 2*std_c_comp_aud_rt)

avg_p_comp_aud_rt = n_affective['comp_aud_RT'].mean()
std_p_comp_aud_rt = n_affective['comp_aud_RT'].std()

p_comp_aud_rt_str = "%3.2f %s %3.2f (ms)" % (avg_p_comp_aud_rt, u"\u00B1", 2*std_p_comp_aud_rt)

#Hit Rates
avg_c_ca_hr = controls['comp_aud_HR'].mean()
std_c_ca_hr = controls['comp_aud_HR'].std()

avg_p_ca_hr = n_affective['comp_aud_HR'].mean()
std_p_ca_hr = n_affective['comp_aud_HR'].std()

c_ca_hr_str = "%3.2f %s %3.2f" % (avg_c_ca_hr, u"\u00B1", 2*std_c_ca_hr)
p_ca_hr_str = "%3.2f %s %3.2f" % (avg_p_ca_hr, u"\u00B1", 2*std_p_ca_hr)
```

```python
#FLUID and CYSTALLIZED INTELLIGENCE
avg_c_fl_int = controls['cogcomp01_nih_fluidcogcomp_unadjusted'].mean()
std_c_fl_int = controls['cogcomp01_nih_fluidcogcomp_unadjusted'].std()

avg_p_fl_int = n_affective['cogcomp01_nih_fluidcogcomp_unadjusted'].mean()
std_p_fl_int = n_affective['cogcomp01_nih_fluidcogcomp_unadjusted'].std()

c_fl_int_str = "%3.2f %s %3.2f" % (avg_c_fl_int, u"\u00B1", 2*std_c_fl_int)
p_fl_int_str = "%3.2f %s %3.2f" % (avg_p_fl_int, u"\u00B1", 2*std_p_fl_int)

avg_c_cry_int = controls['cogcomp01_nih_crycogcomp_unadjusted'].mean()
std_c_cry_int = controls['cogcomp01_nih_crycogcomp_unadjusted'].std()

avg_p_cry_int = n_affective['cogcomp01_nih_crycogcomp_unadjusted'].mean()
std_p_cry_int = n_affective['cogcomp01_nih_crycogcomp_unadjusted'].std()

c_cry_int_str = "%3.2f %s %3.2f" % (avg_c_cry_int, u"\u00B1", 2*std_c_cry_int)
p_cry_int_str = "%3.2f %s %3.2f" % (avg_p_cry_int, u"\u00B1", 2*std_p_cry_int)
```

```python
#EMOTION RECOGNITION (RESPONSE TIME AND CORRECT RESPONSES)
er_c_avg_rt = controls['er4001_er40_c_rtcr'].mean()
er_c_std_rt = controls['er4001_er40_c_rtcr'].std()

er_p_avg_rt = n_affective['er4001_er40_c_rtcr'].mean()
er_p_std_rt = n_affective['er4001_er40_c_rtcr'].std()

er_c_str = "%3.2f %s %3.2f (ms)" % (er_c_avg_rt, u"\u00B1", 2*er_c_std_rt)
er_p_str = "%3.2f %s %3.2f (ms)" % (er_p_avg_rt, u"\u00B1", 2*er_p_std_rt)

#Correct responses
er_c_avg_cr = controls['er4001_er40_c_cr'].mean()
er_c_std_cr = controls['er4001_er40_c_cr'].std()

er_p_avg_cr = n_affective['er4001_er40_c_cr'].mean()
er_p_std_cr = n_affective['er4001_er40_c_cr'].std()

er_c_cr_str = "%3.2f %s %3.2f" % (er_c_avg_cr, u"\u00B1", 2*er_c_std_cr)
er_p_cr_str = "%3.2f %s %3.2f" % (er_p_avg_cr, u"\u00B1", 2*er_p_std_cr)
```

```python
#WASI IQ
avg_c_iq = grouped_df['wasi201_iqscores_full2iq'].mean()[0]
std_c_iq = grouped_df['wasi201_iqscores_full2iq'].std()[0]

c_iq_string = "%3.2f %s %3.2f" % (avg_c_iq, u"\u00B1", std_c_iq)

avg_p_iq = grouped_df['wasi201_iqscores_full2iq'].mean()[2]
std_p_iq = grouped_df['wasi201_iqscores_full2iq'].std()[2]

p_iq_string = "%3.2f %s %3.2f" % (avg_p_iq, u"\u00B1", std_p_iq)

p_iq_string
```

```python
#PANSS SCORES
avg_PANSS_t = n_affective['PANSS_total'].mean()
avg_PANSS_g = n_affective['PANSS_general'].mean()
avg_PANSS_n = n_affective['PANSS_negative'].mean()
avg_PANSS_p = n_affective['PANSS_positive'].mean()

std_PANSS_t = n_affective['PANSS_total'].std()
std_PANSS_g = n_affective['PANSS_general'].std()
std_PANSS_n = n_affective['PANSS_negative'].std()
std_PANSS_p = n_affective['PANSS_positive'].std()

PANSS_t_string = "%3.2f %s %3.2f" % (avg_PANSS_t, u"\u00B1", 2*std_PANSS_t)
PANSS_g_string = "%3.2f %s %3.2f" % (avg_PANSS_g, u"\u00B1", 2*std_PANSS_g)
PANSS_n_string = "%3.2f %s %3.2f" % (avg_PANSS_n, u"\u00B1", 2*std_PANSS_n)
PANSS_p_string = "%3.2f %s %3.2f" % (avg_PANSS_p, u"\u00B1", 2*std_PANSS_p)
```

```python
#chi square test for independence for proportion of male vs female in control vs patient group
n_c_sexes = [28, 11]
n_p_sexes = [45, 12]

observed = [[28,11],[45,12]]

chisq_test_sexes = stats.chi2_contingency(observed)

chisq_sex_pval = chisq_test_sexes[1]
chisq_sex_pval
```

```python
controls = df_red2.loc[df_red2['phenotype_description'] == 0]
n_affective = df_red2.loc[df_red2['phenotype_description'] == 2]
```

```python
#Testing differences in means for IQ, Age, and mother's education level
#age_demographic_test = stats.ttest_ind_from_stats(avg_c_age, std_c_age, num_controls, avg_p_age, std_p_age, num_patients)

age_demographic_test = stats.ttest_ind(controls['interview_age'], n_affective['interview_age'])

IQ_demographic_test = stats.ttest_ind(controls['wasi201_iqscores_full2iq'].dropna(), n_affective['wasi201_iqscores_full2iq'].dropna())
IQ_demographic_pval = round(IQ_demographic_test[1],10)

mot_ed_test = stats.mannwhitneyu(controls['mot_edscale'].dropna(),n_affective['mot_edscale'].dropna())
```

```python
#testing differences in auditory response time and hit rates
comp_aud_rt_test = stats.ttest_ind(controls['comp_aud_RT'].dropna(), n_affective['comp_aud_RT'].dropna())
comp_aud_hr_test = stats.ttest_ind(controls['comp_aud_HR'].dropna(), n_affective['comp_aud_HR'].dropna())

print(comp_aud_rt_test[1], comp_aud_hr_test[1])
```

```python
#testing differences in fluid and crystallized cognition scores
fl_int_test = stats.ttest_ind(controls['cogcomp01_nih_fluidcogcomp_unadjusted'].dropna(), n_affective['cogcomp01_nih_fluidcogcomp_unadjusted'].dropna())

cry_int_test = stats.ttest_ind(controls['cogcomp01_nih_crycogcomp_unadjusted'].dropna(), n_affective['cogcomp01_nih_crycogcomp_unadjusted'].dropna())

print(fl_int_test[1], cry_int_test[1])
```

```python
#testing differences in emotion recognition response time (For correct responses) and number of correct responses
er_rt_test = stats.ttest_ind(controls['er4001_er40_c_rtcr'].dropna(), n_affective['er4001_er40_c_rtcr'].dropna())
er_cr_test = stats.ttest_ind(controls['er4001_er40_c_cr'].dropna(), n_affective['er4001_er40_c_cr'].dropna())
print(er_rt_test, '\n', er_cr_test)
```

```python
summary_stats_df = pd.DataFrame(columns=['Labels','Controls','Patients','P-val', 'df'])   
```

```python
var_names = ['Sex (M/F)','Age (Years)','Mother\'s Education', 'Auditory RT','Auditory HR','Fluid Cognition','Crystallized Cognition','Emotion Recognition (RT)','Emotion Recognition (CR)','Wasi IQ', 'PANSS Positive', 'PANSS Negative', 'PANSS Total', 'PANSS General']
```

```python
def bold_significant_ps(x):
    return 'font-weight: bold' if x <= 0.05 else ''
```

```python
for v, lab in enumerate(var_names):
    summary_stats_df.at[v,'Labels'] = lab

summary_stats_df.at[0,'Controls'] = counts_sex_c
summary_stats_df.at[0,'Patients'] = counts_sex_p
summary_stats_df.at[0,'P-val'] = chisq_sex_pval
summary_stats_df.at[0,'df'] = 1

summary_stats_df.at[1,'Controls'] = control_age_string
summary_stats_df.at[1,'Patients'] = patient_age_string
summary_stats_df.at[1,'P-val'] = age_demographic_test[1]
summary_stats_df.at[1,'df'] = age_demographic_test.df

summary_stats_df.at[2,'Controls'] = mot_c_str
summary_stats_df.at[2,'Patients'] = mot_p_str
summary_stats_df.at[2,'P-val'] = mot_ed_test[1]
summary_stats_df.at[2,'df'] = 1

summary_stats_df.at[3,'Controls'] = c_comp_aud_rt_str
summary_stats_df.at[3,'Patients'] = p_comp_aud_rt_str
summary_stats_df.at[3,'P-val'] = comp_aud_rt_test[1]
summary_stats_df.at[3,'df'] = comp_aud_rt_test.df

summary_stats_df.at[4,'Controls'] = c_ca_hr_str
summary_stats_df.at[4,'Patients'] = p_ca_hr_str
summary_stats_df.at[4,'P-val'] = comp_aud_hr_test[1]
summary_stats_df.at[4,'df'] = comp_aud_hr_test.df

summary_stats_df.at[5,'Controls'] = c_fl_int_str
summary_stats_df.at[5,'Patients'] = p_fl_int_str
summary_stats_df.at[5,'P-val'] = fl_int_test[1]
summary_stats_df.at[5,'df'] = fl_int_test.df

summary_stats_df.at[6,'Controls'] = c_cry_int_str
summary_stats_df.at[6,'Patients'] = p_cry_int_str
summary_stats_df.at[6,'P-val'] = cry_int_test[1]
summary_stats_df.at[6,'df'] = cry_int_test.df

summary_stats_df.at[7,'Controls'] = er_c_str
summary_stats_df.at[7,'Patients'] = er_p_str
summary_stats_df.at[7,'P-val'] = er_rt_test[1]
summary_stats_df.at[7,'df'] = er_rt_test.df

summary_stats_df.at[8,'Controls'] = er_c_cr_str
summary_stats_df.at[8,'Patients'] = er_p_cr_str
summary_stats_df.at[8,'P-val'] = er_cr_test[1]
summary_stats_df.at[8,'df'] = er_cr_test.df
 
summary_stats_df.at[9,'Controls'] = c_iq_string
summary_stats_df.at[9,'Patients'] = p_iq_string
summary_stats_df.at[9,'P-val'] = IQ_demographic_test[1]
summary_stats_df.at[9,'df'] = IQ_demographic_test.df

summary_stats_df.at[10,'Patients'] = PANSS_n_string
summary_stats_df.at[11,'Patients'] = PANSS_p_string
summary_stats_df.at[12,'Patients'] = PANSS_t_string
summary_stats_df.at[13,'Patients'] = PANSS_g_string

summary_stats_df.style \
.format(precision=4, thousands=',', decimal='.') \
.format_index(str.upper, axis=1) \
.applymap(bold_significant_ps, subset=['P-val']) \
.hide()
```
